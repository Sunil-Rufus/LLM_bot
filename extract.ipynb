```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from langchain.embeddings.openai import OpenAIEmbeddings
```

```python
chrome_options = Options()
chrome_options.add_argument("--ignore-ssl-errors=yes")
chrome_options.add_argument("--ignore-certificate-errors")
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(options=chrome_options)
```

```python
file_path = 'unique_urls.txt'  # Replace with the actual path to your text file
urls = []  # Create an empty list to store the lines

with open(file_path, 'r') as file:
    # Read and append each line to the list
    for line in file:
        urls.append(line.strip())
```

```python
urls[0]
```

```python
for url in urls:
    #url = "https://engineering.buffalo.edu/computer-science-engineering/news-and-events.html"
    driver.get(url)
    details = driver.find_element(By.XPATH, '/html/body')
    print(details.text)
    modified_url = url.replace('/', '_')
    #modified_url = modified_url.replace('.html')
    output_file_path = "docs/"+modified_url+".txt"
    with open(output_file_path, "w") as output_file:
        output_file.write(details.text + "\n")
```

```python
name_1 = url.split('/')[-1].split('.')[0]
name_2 = url.split('/')[-2]
if '.' in name_2:
    name_3 = name_2.split('.')[-1]
```

```python
if name_3:
    output_file_path = "docs/"+name_3+"_"+name_1+".txt"
    name_3 = None
    
else:
    output_file_path = "docs/"+name_2+"_"+name_1+".txt"

```
